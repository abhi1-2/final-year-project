```python
import os
main_path='UTKface_Aligned_cropped/crop_part1/'
d=os.listdir(main_path)
count=0
dict={}
for i in range(0,5):
    dict[i]=[]


for i,dirs in enumerate(sorted(d)):
    age=dirs.split("_")[2]
    try:
        dict[int(age)].append(dirs)
    except:
        pass
```

```python
dict_count={}
for key in dict.keys():
    count=0
    for pic in dict[key]:
        count+=1
    dict_count[key]=count
dict_count
```

```python
def age_of_image(pic):
    return int(pic.split("_")[0])

main_dict={}

for key in dict.keys():
    age_arr=[]
    dict_age={}
    arr=[]
    for i,pic in enumerate(dict[key]):
        age=age_of_image(pic)
        if age in age_arr:
            if(dict_age[age]>10):
                pass
            else:
                dict_age[age]+=1
                arr.append(pic)
            
        else:
            age_arr.append(age)
            arr.append(pic)
            dict_age[age]=1
    main_dict[key]=arr
            
```

```python
dict_count={}
for key in main_dict.keys():
    count=0
    for pic in main_dict[key]:
        count+=1
    dict_count[key]=count
dict_count
```
